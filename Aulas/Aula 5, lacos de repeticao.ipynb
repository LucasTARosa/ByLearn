# Lista de todos os alunos  
## Necessidade de uso
Precisamos interagir com vários elementos de uma lista ou grupo de dados, ou com dados de tamanho dinâmico


Ao invés de fazermos igual ao exemplo abaixo, um caso para cada quantidade de dados, podemos fazer de uma só vez

```python
alunos = ['Lucas','Luana','Mateus','Rafaela']

if(len(alunos) == 0):
    print('Não tem nenhum aluno')
elif(len(alunos) == 1):
    print(alunos[0])
elif(len(alunos) == 2):
    print(alunos[0])
    print(alunos[1])
elif(len(alunos) == 3):
    print(alunos[0])
    print(alunos[1])
    print(alunos[2])
elif(len(alunos) == 4):
    print(alunos[0])
    print(alunos[1])
    print(alunos[2])
    print(alunos[3])
```

Pode ser feito da seguinte forma:

```python
alunos = ['Lucas','Luana','Mateus','Rafaela']

for aluno in alunos:
    print(aluno)
```

```python
def main ():
    alunos = ['Lucas', 'Luana', 'Mateus', 'Rafaela']
    notas = [[10,10],[10,9],[7,9],[5,5]]

    for aluno in alunos:
        print(aluno)
    
    for lista_notas in notas:
        soma = 0
        for nota in lista_notas:
            soma += nota
        media = soma / len(lista_notas)
        print(media)

main()
```

```python
def main ():
    alunos = ['Lucas', 'Luana', 'Mateus', 'Rafaela']
    notas = [[10,10],[10,9],[7,9],[5,5]]

    if (len(alunos) == len(notas)):
        for aluno in range(len(alunos)):
            soma = 0
            for nota in notas[aluno]:
                soma += nota
            media = soma / len(notas[aluno])
            print('O aluno (a)', alunos[aluno], 'teve média igual a', media)
            print(f'O aluno (a) {alunos[aluno]} tem a média {media}')
            if media > 6:
                print(f'- Portanto, ele(a) foi aprovado(a)')
            else:
                print(f'- Infelizmente ele(a) foi reprovado(a)')
    else:
        print('Tem algo de errado! \nPossivelmente algum aluno não fez ao menos uma prova')
    

main()
```

**Assim como podemos ter listas dentro de listas, nós podemos ter for dentro de for**  
Na computação chamamos isso de for aninhado  
  
    
**Os laços for (laços de repetição) são muito úteis para se trabalhar com índices**

**Laços de repetição for também funcionam com strings, mostrando caracter por caracter**

```python
string = 'Curso de Python da ByLearn'

for char in string:
    print(char)
```

```python
lista = ['B','y','L','e','a','r','n']

for valor in lista:
    print(valor)
```

```python
for valor in enumerate(lista):
    print(valor)
```

```python
for valor in enumerate(lista):
    for indice in valor:
        print(indice)
```

```python
for chave, valor in enumerate(lista):
    print(chave, valor)
```

**1º nós podemos usar o enumerate() para além do valor, também termos o índice**  
**2º Não precisamos necesssariamente usar for aninhada, podemos usar a vírgula para definir os elementos que queremos recuperar (ex: chave, valor)**

## Outro laço de repetição é o while, que significa enquanto
### Funciona enquanto uma condição for verdadeira fazendo uma determinada ação

```python
contador = 1

while contador < 6: 
    print(contador) # Devemos tomar cuidado para não entrar em loop
```

```python
contador = 1

while contador < 6: 
    print(contador)
    contador += 1
```

```python
condicao_de_parada = '0'

while condicao_de_parada != '9':
    condicao_de_parada = input('Digite um número: ')
    for i in condicao_de_parada:
        print(i)
```

```python
condicao_de_parada = '0'

while condicao_de_parada != '9':
    condicao_de_parada = input('Digite um número: ')
    for i in range(1,int(condicao_de_parada)+1):
        print(i)
```

## Utilizando Continue e Break

**Continue** -> Continuar em português (continuo a minha operação de loop, vou para a próxima etapa)  
**Break** -> Quebra em português (quebrar a minha operação de loop, interromper o loop nesta etapa)  

```python
i = 1

while i < 5:
    print(i)
    i += 1
    if i == 3:
        break
print('---'*10)
for x in range (10,20):
    print(x)
    if x == 15:
        break
```

```python
i = 0
while i < 6:
    i += 1
    print(i)
    if i == 3:
        continue
print('----' * 10)

for x in range (10,20):
    print(x)
    if x == 15:
        continue
    
```

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

## Desafio do par ou impar

Neste desafio você deve mostrar todos os números pares e pular os ímpares.  
No intervalo de 0 a 20, onde se chegar no número 15, nós vamos parar a execução

```python
def init():
    x = 0
    while x < 20:
        x += 1
        if x%2 == 0:
            print(x)
        elif x == 15:
            break
        else:
            continue
        
init()
```

```python
for numero in range(20):
    if numero == 15:
        break
    if numero %2 == 1:
        continue
    print(numero)
```

## Extra: Há o else em laços de repetição!
Esse else ocorre depois do fim do laço

```python
for i in range(5):
    print(i)
else:
    print('Após o laço, após a iteração')
```

```python
i = 1
while i < 5:
    print(i)
    i += 1
else:
    print('i não é mais menor do que 5')
    print(i)
```

## Diferença de uso entre For e While

O laço **for** é geralmente usado quando nós sabemos o número de repetições de antemão;  



Já o **while** é para quando temos uma ideia de condição de para, mas não sabemos exatamente quando isso acontece (número de iterações até isso acontecer)
